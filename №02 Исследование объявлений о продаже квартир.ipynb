### Заполнение пропусков

Посчитаем, сколько в таблице пропущенных значений. Для этого воспользуемся двумя методами pandas (isna() - чтобы определить пустые ячейки и sum() - для подсчета их количества в каждом столбце) :

```python
df.isna().sum().sort_values()
```

Создадим функцию `nan_share(column)`, которая выведет долю в процентах пропущенных значений в столбце:

```python
def nan_share(column):
    return round(((df[column].isna().sum() / len(df)) * 100) , 2)
```

Кажется, что в колонках `floors_total` и `locality_name` доля пропусков незначительная:

```python
print((nan_share('floors_total') + nan_share('locality_name')),'%')
```

— менее одного процента, поэтому ими можем пренебречь. Удалим строки. 

```python
# удалим строки
df = df.dropna(subset=['floors_total','locality_name'])

# убедимся, что пропусков не осталось
df[['floors_total','locality_name']].isna().sum()
```

Рассмотрим остальные колонки с пропущенными значениями.

Заполним пропуски в колонке `ceiling_height_m`. 
Типовые дома, как правило, обладают схожими характеристиками, в том числе и высотой потолков. На основании этого факта будет логичным заменить пропуски медианными значениями по домам одной этажности (столбец `floors_total`).

```python
# создадим сводную таблицу ceiling_medians по этажности домов и соответствующих медианных значений высоты потолков
ceiling_medians = df.pivot_table(index='floors_total', values='ceiling_height_m', aggfunc=['median', 'count'])
ceiling_medians.columns = ['median', 'count']
ceiling_medians
```

```python
# на основании данных из ceiling_medians заполним пропуски в ceiling_height_m
for floor in ceiling_medians.index:
    df.loc[(df['ceiling_height_m'].isna()) & 
             (df['floors_total'] == floor), 
             'ceiling_height_m'] = ceiling_medians.loc[floor, 'median']
    
df['ceiling_height_m'].isna().sum()
```

Осталось одно пропущенное значение. Это произошло из-за того, что нам встретилась всего одна квартира из дома в котором 33 этажа. Будем считать, что в этом доме такая же высота потолков, как и у квартиры в 34-этажом здании. 

```python
# заменим пропущенную высоту потолков в квартире 33-этажного здания на высоту потолков в квартире 34-этажного дома
df.loc[df['floors_total'] == 33, 'ceiling_height_m'] = ceiling_medians.loc[34, 'median']

# убедимся, что в столбце ceiling_height_m больше нет пропусков
df['ceiling_height_m'].isna().sum()
```

Разберемся с пропусками в столбце `days_exposition`: посмотрим, как влияют на пропуски даты размещения объявлений.

```python
# изменим формат даты на корректный, без времени
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format = '%Y-%m-%d')

# построим гистограмму, отражающую пропуски по годамм 
df[df['days_exposition'].isna()]['first_day_exposition'].dt.year.hist()

plt.title('Частота встречаемых пропущенных значений в days_exposition по годам')
```

Из графика видно, что чем "свежее" объявления, тем больше пропущенных значений в столбце `days_exposition`. Вероятнее всего, это вовсе не пропуски. Отсутствие значения говорит о том, что объявление не было снято с публикации и еще актуально. В таком случае, NaN'ы логично заменить нулями.

```python
df['days_exposition'] = df['days_exposition'].fillna(0)
df['days_exposition'].isna().sum()
```

Обратимся к пропускам в колонке `city_centers_nearest_m`, посмотрим, какую долю от всех пропусков составляют пропуски по Питеру: 

```python
round(len(df[(df['locality_name'] == 'Санкт-Петербург') & 
             (df['city_centers_nearest_m'].isna())]) / len(df.query('locality_name == "Санкт-Петербург"')), 3)
```

Узнаем, есть ли пропуски в столбце `airports_nearest_m` по Питеру:

```python
len(df.loc[(df['airports_nearest_m'].isna()) & (df['locality_name'] == 'Санкт-Петербург')])
```

Пропуски в удаленности от центра и аэропорта для Санкт-Петербурга практически не встречаются. Значит основная доля пропусков встречается в других населенных пунктах, а, учитывая, что бóльшая часть объявлений - квартиры в Питере, мы не можем заполнить их медианным значением всех расстояний. Поэтому заполним пропуски медианными значениями по двум группам: "медаиана для Питера" и "медиана для остальных городов". Сделаем это с помощью функции `fillna_median_by_group`:

```python
def fillna_median_by_group(column_1, column_2):

    for type in df[column_2].unique():
        if type == 'Санкт-Петербург':
            median = df.loc[df[column_2] == 'Санкт-Петербург', column_1].median()
            df.loc[(df[column_1].isna()) & (df[column_2] == 'Санкт-Петербург'), column_1] = median
        else:
            median = df.loc[df[column_2] != 'Санкт-Петербург', column_1].median()
            df.loc[(df[column_1].isna()) & (df[column_2] != 'Санкт-Петербург'), column_1] = median
        
    
fillna_median_by_group('city_centers_nearest_m', 'locality_name')
fillna_median_by_group('airports_nearest_m', 'locality_name')
df[['city_centers_nearest_m','airports_nearest_m']].isna().sum()
```

Очевидно, что пропущенные значения в столбцах `airports_nearest_m`, `city_centers_nearest_m` появились в результате технических ошибок, так как эти значения заполняются автоматически. На эту проблему следует указать команде Яндекс Недвижимость.

А высоту потолков - `ceiling_height_m` пользователи могли не указать, либо потому что это - не обязательное для заполнения поле, либо просто забыв, или не знав.

Перейдем к балконам. Вероятнее всего, пропуски в колонке `number_of_balconies` появились потому, что пользователи не стали указывать их количество, так как их просто нет. Здесь логично заполнить отсутствующие значения нулями.

```python
# заполним нулями и убедимся, что пропусков не осталось
df['number_of_balconies'].fillna(0, inplace = True)
df['number_of_balconies'].isna().sum() 
```

В столбце `is_apartment` пропусков больше всего. Это не удивительно, ведь апартаменты гораздо реже встречаются на рынке. Пропущенные значения в этой колонке, скорее всего - незаполненные пользователем поля, которые означают, что помещения жилые. Заполним пропуски на False.

```python
# заменим NaN на False и убедимся, что пропусков не осталось
df['is_apartment'] = df['is_apartment'].fillna(False)
df['is_apartment'].isna().sum()
```

Избавимся от пропусков в `living_area_m2`. Кажется, что размер жилой площади напрямую зависит от размера самой квартиры, проверим эту гипотезу с помощью диаграммы рассеяния:

```python
# Строим диаграмму рассеивания
df.plot(x='total_area_m2', y='living_area_m2', kind='scatter')
plt.title('Зависимость жилой площади от общей площади квартиры')
plt.xlabel('Общая площадь')
plt.ylabel('Жилая площадь')
plt.show()

# Считаем Коэффициент Пирсона
print('Коэффициент Пирсона',round(df['living_area_m2'].corr(df['total_area_m2']),3))
```

Коэффициент корреляции Пирсона практически равен единице, что говорит о тесной линейной взаимосвязи данных показателей. Учитывая, что в столбце `total_area_m2` нет пропусков, мы можем заполнить пропуски в столбце `living_area_m2`, посчитав среднюю долю жилой площади от общей. Обозначим ее `mean_living_area_share` и заполним пропущенные значения:

```python
# посчитаем среднюю долю жилой площади от общей
mean_living_area_share = (df['living_area_m2'] / df['total_area_m2']).mean()

# заполним пропуски 
df['living_area_m2'] = df['living_area_m2'].fillna(df['total_area_m2'] * mean_living_area_share)

# проверим, что их не осталось
df['living_area_m2'].isna().sum()
```

Проверим аналогичную гипотизу о линейной зависимости площади кухни от нежилой площади квартиры:

```python
# посчитаем размер нежилой площади в отдельном столбце
df['non_residential_area'] = df['total_area_m2'] - df['living_area_m2']

# построим диаграмму рассеивания
df.plot(x='non_residential_area', y='kitchen_area_m2', kind='scatter')
plt.title('Зависимость площади кухни от нежилой площади квартиры')
plt.xlabel('Нежилая площадь')
plt.ylabel('Площадь кухни')
plt.show()

# посчитаем Коэффициент Пирсона
print('Коэффициент Пирсона',round(df['kitchen_area_m2'].corr(df['non_residential_area']),3))
```

Здесь выбросов больше, однако также наблюдается тесная линейная зависимость. Посчитаем среднюю долю кухни от нежилой площади и аналогичным оразом заполним пропуски в `kitchen_area_m2`:

```python
# посчитаем среднюю долю кухни от нежилой площади
mean_kitchen_area_share = (df['kitchen_area_m2'] / (df['total_area_m2'] - df['living_area_m2'])).mean()

# заполним пропуски 
df['kitchen_area_m2'] = df['kitchen_area_m2'].fillna(df['non_residential_area'] * mean_kitchen_area_share)

# проверим, что их не осталось
df['kitchen_area_m2'].isna().sum()
```

Исходя из того, что в общей площади квартир нет пропусков, можно сделать вывод о том, что это - обязательное поле для заполнения, в то время как жилая площадь и площадь кухни - не обязательные и, вероятнее всего, причиной пропусков в этих столбцах является человеческий фактор.

### Расчёты и добавление результатов в таблицу

Добавим в датафрейм дополнительные данные, которые пригодятся в дальнейшем анализе:
- цену квадратного метра;
- день недели, месяц и год публикации объявления;
- этаж квартиры (варианты — первый, последний, другой);
- соотношение жилой и общей площади;
- отношение площади кухни к общей.

```python
# посчитаем цену квадратного метра и приведем к типу int
df['price_per_m2'] = (df['last_price'] / df['total_area_m2']).astype('int')


# создим столбец дня недели
df['weekday'] = df['first_day_exposition'].dt.weekday

# создадим словарь, чтобы заменить цифры на названия
dict_of_weekdays = {1: 'понедельник', 2: 'вторник', 3: 'среда', 4: 'четверг', 
                    5: 'пятница', 6: 'суббота', 7: 'воскресенье'}
# поменяем значения
df['day'] = df['weekday'].map(dict_of_weekdays)


#создадим столбец месяца
df['month'] = df['first_day_exposition'].dt.month

# создадим словарь, чтобы заменить цифры на названия
dict_of_months = {1: 'январь', 2: 'февраль', 3: 'март', 4: 'апрель', 
                    5: 'май', 6: 'июнь', 7: 'июль', 8: 'август', 
                    9: 'сентябрь', 10: 'октябрь', 11: 'ноябрь', 12: 'декабрь'}
# поменяем значения
df['month'] = df['month'].map(dict_of_months)


#создаю столбец года
df['year'] = df['first_day_exposition'].dt.year


#напишем функцию категоризации по этажам 
def floor_category(row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    elif 1 < floor < floors_total:
        return 'другой'
    
#категоризуем этажи с помощью функции  
df['floor_category'] = df.apply(floor_category, axis = 1)

#добавим столбец соотношения жилой площади к общей, округлим до тысячных
df['living_area_ratio'] = round(df['living_area_m2'] / df['total_area_m2'], 3)

#добавим столбец отношения площади кухни к общей, округлим до тысячных
df['kitchen_area_ratio'] = round(df['kitchen_area_m2'] / df['total_area_m2'], 3)

# убедимся, что появились новые столбцы
df.head()
```

### Уберем редкие и выбивающиеся значения.

Рассмотрим выбросы в `total_area_m2`, `last_price`, `rooms`, `ceiling_height_m`:

```python
df['total_area_m2'].min()
```

```python
# получим числовое описание данных по нужным столбцам 
display(df[['total_area_m2', 'last_price', 'rooms', 'ceiling_height_m']].describe())

for col in ['total_area_m2', 'last_price', 'rooms', 'ceiling_height_m', 'days_exposition']:
    df.boxplot(column=col)
    plt.show()

# зададим функцию, получающую на входе название столбца, и возвращающую границы "усов"
def det_mustache(column):
    
    # определим первый и третий квантиль и размах
    q1 = df[[column]].quantile(.25)[column]
    q3 = df[[column]].quantile(.75)[column]

    iqr = q3 - q1
    
    # левый и правый "ус"    
    left_mustache = round(q1 - 1.5 * iqr, 2)
    right_mustache = round(q3 + 1.5 * iqr, 2)

    # левый "ус" не должен быть меньше минимального значения
    if left_mustache < df[column].min(): left_mustache = df[column].min() 
        
    # правый "ус" не должен быть больше максимального значения
    if right_mustache > df[column].max(): right_mustache = df[column].max()
        
    return [left_mustache, right_mustache]

# зададим функцию, вычисляющую долю аномальных значений
def print_anomalies_ratio(column):
    return 'Аномальные значения составляют {:.1%}'. format(len(df.loc[df[column] > det_mustache(column)[1]]) / len(df)) + ' от всей выборки'

```

```python
# выведем границы усов и долю аномальных значений для нужных столбцов
for col in ['total_area_m2', 'last_price', 'rooms', 'ceiling_height_m', 'days_exposition']:
    print(col)
    print('Границы "усов":', det_mustache(col))
    print(print_anomalies_ratio(col))
```

Выбросы присутствуют во всех столбцах и находятся в пределах 9% от всей выборки каждый. Добавим Создадим отдельный датафрейм `good_data` с чистыми данными.

```python
# добавим булевые столбцы для выбивающихся значений
df['too_large_area'] = df['total_area_m2'] > det_mustache('total_area_m2')[1]
df['too_expensive'] = df['last_price'] > det_mustache('last_price')[1]
df['too_many_rooms'] = df['rooms'] > det_mustache('rooms')[1]
df['unusual_ceiling'] = (df['ceiling_height_m'] > 
                           det_mustache('ceiling_height_m')[1]) | (df['ceiling_height_m'] 
                                                                   < det_mustache('ceiling_height_m')[0])
df['too_slow'] = df['days_exposition'] > det_mustache('days_exposition')[1]
df['too_fast'] = df['days_exposition'] < 30

# опираясь на них создадим датафрейм с чистыми данными
good_data = df[(df['too_large_area'] == False) & 
     (df['too_expensive'] == False) &
     (df['too_many_rooms'] == False) &
     (df['unusual_ceiling'] == False) &
     (df['too_slow'] == False) &
     (df['too_fast'] == False)].reset_index(drop=True)

print('Доля чистых значений составляет: {:.1%}'. format(len(good_data) / len(df)))
```

### Посмотрим, как меняется цена в зависимости от удаленности от центра и определим центральную зону.

```python
# создадим столбец в df с удаленностью от центра в км
df['city_centers_nearest_km'] = df['city_centers_nearest_m'].apply(lambda x: x // 1000).astype('int')

# создаем датафрейм по Питеру
St_Petersburg = df.query('locality_name == "Санкт-Петербург"')

# создаем сводную таблицу с медианной ценой квадратного метра
cityCenters_nearest_km_pivot_table = St_Petersburg.pivot_table(index = 'city_centers_nearest_km', values = 'price_per_m2', aggfunc = ['mean'])

# строим график
cityCenters_nearest_km_pivot_table.plot(y = ['mean'],kind = 'bar',figsize = (10,5))
plt.xlabel('Удаленность от центра в километрах')
plt.ylabel('Цена квадратного метра')
```

Центром Питера обозначим иинтервал в 8 км, так как на этой отметке видно явное снижение цены квадратного метра. Как мы заметили ранее: чем дальше от центра - тем ниже цена. Наблюдаются 2 пиковых значения в радиусе 20 и 27 км от центра города.
